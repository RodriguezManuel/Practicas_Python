**Nota de resolución:** Para simplificar los cálculos y evitar problemas al sumar minutos y segundos, la que hice fue convertir todas las duraciones a **segundos totales**. Esto facilitó encontrar la canción más larga, la más corta y la duración total.

```python
playlist = [
    {"title": "Bohemian Rhapsody", "duration": "5:55"},
    {"title": "Hotel California", "duration": "6:30"},
    {"title": "Stairway to Heaven", "duration": "8:02"},
    {"title": "Imagine", "duration": "3:07"},
    {"title": "Smells Like Teen Spirit", "duration": "5:01"},
    {"title": "Billie Jean", "duration": "4:54"},
    {"title": "Hey Jude", "duration": "7:11"},
    {"title": "Like a Rolling Stone", "duration": "6:13"},
]


time_accumulator = [
    0,
    0,
]  # Acumulador para contar los minutos y segundos totales de la playlist
max_duration = -1
max_song = (
    ""  # Variable para almacenar la información de la canción con la duración máxima
)
min_duration = 9999
min_song = (
    ""  # Variable para almacenar la información de la canción con la duración mínima
)
for song in playlist:
    duration = song["duration"].split(":")  # Separa los minutos y segundos
    minutes = int(duration[0])  # Convierte los minutos a entero
    seconds = int(duration[1])  # Convierte los segundos a entero
    total_seconds = minutes * 60 + seconds  # Calcula la duración total en segundos
    if (
        max_duration < total_seconds
    ):  # Compara la duración total en segundos con la duración máxima actual
        max_duration = total_seconds
        max_song = song
    if (
        min_duration > total_seconds
    ):  # Compara la duración total en segundos con la duración mínima actual
        min_duration = total_seconds
        min_song = song
    time_accumulator[0] += minutes  # Suma los minutos al acumulador
    time_accumulator[1] += seconds
if time_accumulator[1] >= 60:  # Si los segundos acumulados superan los 60
    time_accumulator[0] += time_accumulator[1] // 60
    time_accumulator[1] = time_accumulator[1] % 60
print(
    f"Duración total de la playlist: {time_accumulator[0]} minutos y {time_accumulator[1]} segundos"
)
print(f'Canción más larga: "{max_song["title"]}" ({max_song["duration"]})')
print(f'Canción más corta: "{min_song["title"]}" ({min_song["duration"]})')

```
