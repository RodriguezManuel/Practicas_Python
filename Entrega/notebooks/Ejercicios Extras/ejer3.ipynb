**Nota de resolución:** Se procesaron las cadenas de texto (los posteos) para identificar y contar la aparición específica del carácter '#' y extraer las palabras asociadas.

```python
review = """La película sigue a un grupo de astronautas que
viajan a Marte
en una misión de rescate. El capitán Torres lidera al equipo
a través
de tormentas solares y fallos en el sistema de navegación. Al
llegar
a Marte descubren que la base está abandonada y los
suministros
destruidos. Torres decide sacrificar la nave nodriza para
salvar
al equipo y logran volver a la Tierra en una cápsula de
emergencia.
El final revela que Torres sobrevivió gracias a un pasaje
secreto."""


# Pedimos al usuario las palabras separadas por coma
user_input = input("Ingrese las palabras spoiler (separadas por coma): ")


# Separamos la cadena de texto en una lista
spoilers_list = user_input.split(",")


#  Recorremos cada palabra que el usuario ingresó
for word in spoilers_list:
    spoiler = word.strip()  # .strip() elimina espacios en blanco sobrantes a los lado
    if len(spoiler) == 0:  # Si quedó vacío, lo salteamos
        continue
    asterisks = "*" * len(spoiler) # Creo una cadena de astericos del largo de la pala
    spoiler_lower = spoiler.lower() 
    new_review = ""
    i = 0
    # Recorremos todo el texto caracter por caracter
    while i < len(review):
        # Extraigo un pedacito del texto del mismo tamaño que el spoiler actual
        text_chunk = review[i : i + len(spoiler)] 


        # Comparamos ignorando mayúsculas/minúsculas usando .lower()
        if text_chunk.lower() == spoiler_lower:
            new_review += asterisks
            i += len(
                spoiler
            )  # Damos un "salto" para no volver a leer lo que ya reemplazamos
        else:
            new_review += review[i]
            i += 1


    # Actualizamos el texto oficial para la siguiente palabra de la lista
    review = new_review


# 5. Imprimimos el resultado final
print(review)

```
