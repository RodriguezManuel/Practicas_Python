# Ejercicio 4: 

**Nota de resolución:** Se validaron los strings comprobando que tengan exactamente un '@', extrayendo el `username` y el `domain` con un `.split()`, y verificando que no empiecen ni terminen con caracteres inválidos (como '@' o '.').

```python
email = input("Ingrese un correo electrónico: ")
is_valid = True  # Asumo que el correo es valido


# Validamos el primer criterio: debe contener exactamente un @
if email.count("@") != 1:
    is_valid = False
else:
    # Si tiene exactamente un @, podemos dividir el correo en dos mitades
    parts = email.split(
        "@"
    )  # Esto lo que hace es devolver una lista con dos portes la parte antes
    username = parts[0]  # Lo que está antes del @
    domain = parts[1]  # Lo que está después del @


    # 2. No empieza ni termina con @ ni con .
    # Verificamos los extremos de la cadena original completa
    if (
        email.startswith("@")
        or email.startswith(".")
        or email.endswith("@")
        or email.endswith(".")
    ):
        is_valid = False


    # 3. Tiene al menos un carácter antes del @
    elif len(username) < 1:
        is_valid = False


    # 4. Tiene al menos un punto (.) después del @
    elif "." not in domain:
        is_valid = False


    # 5. La parte después del último punto tiene al menos 2 caracteres
    else:
        # Divimos la parte del dominio usando el punto como separador
        domain_parts = domain.split(".")


        # Con el -1 agarro la ultima parte del dominio, que es la que debe  t
        last_part = domain_parts[-1]
        if len(last_part) < 2:
            is_valid = False
# Resultado
if is_valid:
    print("El correo electrónico es VÁLIDO.")
else:
    print("El correo electrónico es INVÁLIDO.")

```
