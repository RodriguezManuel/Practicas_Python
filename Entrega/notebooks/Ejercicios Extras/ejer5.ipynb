**Nota de resolución:** En lugar de usar una cadena larguísima de `if/elif`, se implementó una tabla de búsqueda  utilizando un diccionario. Esto hace que buscar el precio según el peso y la zona sea mucho más directo, limpio y eficiente.

```python
shipping_prices = {
    "local": {"Liviano": 500, "Mediano": 1000, "Pesado": 2000},
    "regional": {"Liviano": 1000, "Mediano": 2500, "Pesado": 5000},
    "nacional": {"Liviano": 2000, "Mediano": 4500, "Pesado": 8000},
}
zone = (
    input(f"Ingrese la zona de destino (local/regional/nacional): ").lower()
)  # Primero ingreso la zona asi, me aseguro de leer el peso solo si la zona es valida
if zone not in shipping_prices:
    print("Zona no válida. Las zonas disponibles son: local, regional o nacional.")
else:
    weight = float(
        input(f"Ingrese el peso del paquete (kg): ")
    )  # Ingresa el peso del paquete
    if weight < 1:
        print(f"El costo de envío es: ${shipping_prices[zone]['Liviano']}")
    elif weight >= 1 and weight < 5:
        print(f"El costo de envío es: ${shipping_prices[zone]['Mediano']}")
    else:
        print(f"El costo de envío es: ${shipping_prices[zone]['Pesado']}")
```
