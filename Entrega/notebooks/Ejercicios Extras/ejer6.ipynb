**Nota de resolución:** Se validó que el ingreso tenga al menos 3 participantes y no haya nombres duplicados (ignorando mayúsculas y espacios). Para el sorteo aleatorio, se utilizó `random.shuffle()` y el método de desplazamiento con módulo (`%`) para asegurar que nadie se saque a sí mismo.

```python
from collections import Counter


posts = [
    "Arrancando el lunes con energía #Motivación #NuevaSemana",
    "Terminé mi primer proyecto en Python #Python #Programación #OrgullosoDeMi",
    "No puedo creer el final de la serie #SinSpoilers #SerieAdicta",
    "Nuevo video en el canal sobre #InteligenciaArtificial y #Python",
    "Entrenamiento de hoy completado #Fitness #Motivación #NoPainNoGain",
    "Leyendo sobre #InteligenciaArtificial y el futuro del trabajo #Tecnología",
    "Arranqué a estudiar #Programación por mi cuenta #Python #Autodidacta",
    "Finde de lluvia, maratón de series #SerieAdicta #Relax",
    "Workshop de #InteligenciaArtificial en la universidad #Tecnología #Programación",
]
hashtasg_lista = []


# Primer parte extraigo todos los hashtags de los posts y los agrego a una lista
# Lo que hago es recorrer cada post, luego separar cada post en palabras, y por cada palabra verifico si es un hashtag (si empieza con #) y si lo es lo agrego a la lista de hashtags
for post in posts:
    words = post.split()  # Separa el post en palabras
    for word in words:
        if word.startswith("#"):  # Si la palabra es un hashtag
            hashtasg_lista.append(word)  # Lo agrego a la lista de hashtags
# Cuento la frecuencia de cada uno de los hashtags usando Counter
hashtag_counts = Counter(hashtasg_lista)
unique_hashtags = 0
# Muestro los hastagas que tiene mas de una ocurrencia
print("Hashtags trending (más de una aparición):")
for hashtag, count in hashtag_counts.most_common():
    if count > 1:  # Si el hashtag tiene mas de una ocurrencia
        print(f"\t{hashtag}: {count} veces") # El \t tabula y deja la imprimesion como esta en el enunciado
print(f"Cantidad de hashtags únicos: {len(hashtag_counts)}")

```
