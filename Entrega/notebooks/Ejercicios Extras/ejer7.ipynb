```python
import random


names = input("Ingrese los participantes (separados por coma): ")


# 1. Separamos por coma
dirty_names = names.split(
    ","
)  # Esto devuelve una lista con cada nombre, pero con espacios al principio o al final, por eso se l
participants = []
# 2. Limpiamos con .strip() y validamos duplicados
for n in dirty_names:
    # Usamos .strip() para sacar los espacios que quedan al escribir "Ana, Pedro"
    clear_name = n.strip()


    if clear_name != "":
        # Chequeamos si el nombre ya está (ignorando mayúsculas)
        ready = False
        for p in participants:  # Recorro la lista de participantes ya agregados y
            if p.lower() == clear_name.lower():
                ready = True


        if not ready:
            participants.append(clear_name)


# 3. Validación y Sorteo
if len(participants) < 3:
    print("Error: Se necesitan al menos 3 participantes.")
else:
    random.shuffle(
        participants
    )  # Esto mezcla la lista de participantes de forma aleatoria, asi el sorteo es diferente cada ve


    print("\nSorteo de amigo invisible:")
    for i in range(len(participants)):
        # El último le regala al primero, los demás al que sigue
        id_receptor = (i + 1) % len(participants)
        print(f" {participants[i]} → {participants[id_receptor]}")

```
