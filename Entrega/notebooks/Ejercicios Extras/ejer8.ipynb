```python
message = input("Ingrese un mensaje: ")
value = int(input("Ingrese un valor de desplazamiento (número entero): "))
encrypted_message = ""
z_value = ord(
    "z"
)  # Obtenemos el código ASCII de la letra 'z' para manejar el caso de las letras al final del alfabeto
Z_value = ord(
    "Z"
)  # Obtenemos el código ASCII de la letra 'Z' para manejar el caso de las letras mayúsculas al final del alfabeto


for char in message:
    if char.isalpha():  # Si es una letra
        if char.islower():  # Camino 1: ES MINÚSCULA
            if ord(char) + value > z_value:
                new_char = chr((ord(char) + value - z_value) + ord("a") - 1)
            else:
                new_char = chr(ord(char) + value)
        elif char.isupper():  # Camino 2: ES MAYÚSCULA
            if ord(char) + value > Z_value:
                new_char = chr((ord(char) + value - Z_value) + ord("A") - 1)
            else:
                new_char = chr(ord(char) + value)
        encrypted_message += new_char
    else:  # Camino 3: NO ES UNA LETRA (espacios, números, etc.)
        encrypted_message += char  # Lo agregamos tal cual, sin cambios
print("Mensaje encriptado:", encrypted_message)

```
