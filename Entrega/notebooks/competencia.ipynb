Se llamaran a funciones para que el codigo se vea mas legible, poder segmentar los problemas a resolver (dividirlo en partes) me ayuda a identificar los errores logicos mas rapido.

```python
for round_actual in rounds:
    #  Llamamos a la función y GUARDAMOS lo que nos devuelve
    scores, round_winner, highest_score = beginning_round(round_actual, number_of_rounds)
    
    for particpant, jugdges in scores.items():
        round_score = sum(jugdges.values())
        Standings[particpant]["Puntaje"] += round_score
        
        # Llamamos a la función que actualiza los máximos 
        highest_score, round_winner = score_max(round_score, highest_score, particpant, round_winner)
        
        # 3. Llamamos a la función que actualiza el diccionario directamente
        update_maxscore(Standings, round_score, particpant)
        
    # Fuera del for de participantes, actualizamos las victorias
    Standings[round_winner]["Rondas ganadas"] += 1
    
    # Imprimimos el ganador
    print(f"Ronda {number_of_rounds} - {round_actual['theme']}: \n Ganador: {round_winner} ({highest_score} pts)")
    print("-" * 30) # Cadena de guiones para separar cada ronda 
    
    number_of_rounds += 1 # Sumamos 1 a la ronda para la siguiente iteracion 
```
